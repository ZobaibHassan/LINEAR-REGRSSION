```python
import pandas as pd
import numpy as np
```

```python
    data = pd.read_csv('assignment1.csv')
    print(data.describe())
    X = data['Age (years)']
    Y = data['Salary (PKR)']
    mean_x = np.mean(X)
    mean_y = np.mean(Y)
    size = len(X)

    numerator = 0
    denominator = 0
    for i in range(size):
        numerator += (X[i] - mean_x)*(Y[i] - mean_y)
        denominator += (X[i]-mean_x) ** 2
    m = numerator/denominator
    c = mean_y - (m * mean_x)
    c

```

```python

max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 1000)
y =  m*x + c

import matplotlib.pyplot as plt

plt.plot(x, y, color='Blue', label='Regression Line')
plt.scatter(X, Y, c='#ef5423', label='Scatter Plot')

plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show

```
